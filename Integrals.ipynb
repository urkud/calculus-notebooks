```python
import sympy as sp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
sp.init_printing()
%matplotlib inline
```

```python
def plot_RS(a, b, n, N, f, method):
    x = np.linspace(a, b, N + 1)
    y = f(x)
    X = np.linspace(a, b, n * N + 1)
    Y = f(X)
    plt.plot(X, Y, 'b')
    x_test = x
    y_test = y
    align = 'edge'
    width = (b - a) / N
    if method == "left":
        x_test = x[:-1]
        y_test = y[:-1]
    elif method == "mid":
        x_test = (x[:-1] + x[1:]) / 2
        y_test = f(x_test)
        align = 'center'
    elif method == "right":
        x_test = x[1:]
        y_test = y[1:]
        width = -width
    plt.plot(x_test, y_test, 'b.', markersize=10)
    plt.bar(x_test, y_test, width=width, alpha=0.3, align=align, edgecolor='b')
    return sum(y_test * (b - a) / N)

n = 10; N = 2
f = lambda x : x + 1
plt.figure(figsize=(35, 10))
plt.subplot(2, 3, 1)
s = plot_RS(0, 2, n, N, f, "left")
plt.title(f"Left Riemann Sum, N = {N}")
plt.subplot(2, 3, 2)
s = plot_RS(0, 2, n, N, f, "mid")
plt.title(f"Midpoint Riemann Sum, N = {N}")
plt.subplot(2, 3, 3)
s = plot_RS(0, 2, n, N, f, "right")
plt.title(f"Right Riemann Sum, N = {N}")
N = 10
plt.subplot(2, 3, 4)
s = plot_RS(0, 2, n, N, f, "left")
plt.title(f"Left Riemann Sum, N = {N}")
plt.subplot(2, 3, 5)
s = plot_RS(0, 2, n, N, f, "mid")
plt.title(f"Midpoint Riemann Sum, N = {N}")
plt.subplot(2, 3, 6)
s = plot_RS(0, 2, n, N, f, "right")
plt.title(f"Right Riemann Sum, N = {N}")
```
